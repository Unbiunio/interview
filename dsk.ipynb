 endpoint de categorías para obtener la lista de categorías disponible

```python
import requests
import json

# Llamada al endpoint de categorías
response_categories = requests.post('https://es-mycooktouch.group-taurus.com/categories/')
categories_data = response_categories.json()

print("Categorías:", json.dumps(categories_data, indent=4))
```

 endpoint de recetas para obtener un conjunto de recetas. Experimenta con los parámetros limit, skip, y sortBy

```python
params = {
    'profile': 'desktopList',
    'limit': '100',  # Ajusta este valor según lo que necesites
    'skip': '0',
    'sortBy': 'recent',
}
response_recipes = requests.post('https://es-mycooktouch.group-taurus.com/recipes/', params=params)
recipes_data = response_recipes.json()

print("Recetas:", json.dumps(recipes_data, indent=4))
```
