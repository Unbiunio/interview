```python
import pandas as pd

import pandas as pd

def drop_constant_columns(df):
    """
    Elimina las columnas del DataFrame que tienen un único valor único.
    
    Parámetros:
        df (pd.DataFrame): El DataFrame de entrada.
    
    Retorna:
        pd.DataFrame: Un nuevo DataFrame sin las columnas constantes.
    """
    # Encontrar columnas con un único valor único
    constant_columns = [col for col in df.columns if df[col].nunique() == 1]
    
    # Eliminar las columnas constantes
    df = df.drop(columns=constant_columns)
    
    return df

```

```python
# Aplicar la función a df_categorias
df_categorias = drop_constant_columns(df_categorias)


df_categorias
```

```python
import requests
import pandas as pd
import json

# URL del endpoint de recetas
url = "https://es-mycooktouch.group-taurus.com/recipes/"

# Parámetros para la solicitud
params = {
    'profile': 'desktopList',
    'limit': 1000,  # Limitar a 1000 recetas
    'skip': 0,      # Saltar 0 recetas (empezar desde el principio)
    'sortBy': 'recent'  # Ordenar por recetas más recientes
}

# Realizar la solicitud POST
response = requests.post(url, params=params)

# Verificar si la solicitud fue exitosa
if response.status_code == 200:
    print("Solicitud exitosa!")
    data = response.json()  # Convertir la respuesta a JSON
else:
    print(f"Error: {response.status_code}")

# Extraer información de las recetas
recipes = []
for recipe in data['result']:
    recipe_info = {
        "title": recipe["title"],
        "category": recipe["category"],
        "difficulty": recipe["difficulty"],
        "totalTime": recipe["totalTime"],
        "user": recipe["user"]["displayName"],
        "ingredients": [ingredient["name"] for ingredient in recipe["ingredients"]],
        
    }
    recipes.append(recipe_info)

# Convertir la lista de recetas en un DataFrame
df_ingredientes= pd.DataFrame(recipes)

# Mostrar el DataFrame
df_ingredientes
```

```python

# Obtener todos los ingredientes únicos
all_ingredients = set(ingredient for sublist in df_ingredientes['ingredients'] for ingredient in sublist)

# Crear las columnas para los ingredientes
for ingredient in all_ingredients:
    df_ingredientes[ingredient] = df_ingredientes['ingredients'].apply(lambda x: 1 if ingredient in x else 0)

# Eliminar la columna 'ingredients' ya que ahora tenemos las columnas individuales
df_ingredientes = df_ingredientes.drop(columns=['ingredients'])

# Contar cuántas veces aparece cada ingrediente (cuántos 1 hay por columna)
ingredient_counts = df_ingredientes.iloc[:, 5:].sum().sort_values(ascending=False)

# Reordenar las columnas de ingredientes según el conteo (de más a menos)
ordered_columns = ['title', 'category', 'difficulty', 'totalTime', 'user'] + ingredient_counts.index.tolist()

# Reordenar el DataFrame
df_ingredientes = df_ingredientes[ordered_columns]

# Mostrar el DataFrame resultante
df_ingredientes
```
